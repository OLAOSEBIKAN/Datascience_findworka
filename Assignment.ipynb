```python
def print_multiple_of_three(a):
    if a % 3 == 0:
        print(f"{a} is a multiple of 3")
    else: 
        print(f"Sorry {a} is not a multiple of 3")
        
        
print_multiple_of_three(12)
print_multiple_of_three(7)
```

```python
def even_odd():
    print("This function is for getting the even and odd number between 1 to 1000")
    #get range of numbers
    to_1k = range(1,1001)
    for num in to_1k:
        #check if the number is even of odd
        if num % 2 == 0:
            #if num is even do this
            print(f"{num}: even")
        else:
            #if number is odd do this
            print(f"{num}: odd")        
even_odd()
```
